```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import rcParams
from matplotlib import cm, colors
from matplotlib.colors import ListedColormap
import seaborn as sns

from pylab import rc
rc('text',usetex=True)

# %matplotlib inline
# import matplotlib_inline
# matplotlib_inline.backend_inline.set_matplotlib_formats('png')

#matplotlib.rcParams['font.serif'] = ['Computer Modern']
#matplotlib.rcParams['font.family'] = 'serif'


sns.set_theme()
sns.set_style("whitegrid")
sns.set_context("paper")

sns.set_style('ticks')
```

```python
colors = cm.plasma(np.linspace(0, 1, 12))

for i, color in enumerate(colors):
    plt.plot([0, 1], [i, i], color=color)

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (5,6))

x = np.arange(1, 2.1, 0.01)
y1 = x

ax.grid(False, which="major", alpha=0.3)
ax.plot([0, 2], [1, 1], color='black', linewidth=1)
ax.plot([1, 2], [1, 2], color='black', linewidth=1)
ax.plot([1, 1], [0, 2], color='black', alpha = 0.5, linewidth=1,linestyle='dashed')

ax.fill_between(x, y1, 1, color=colors[7], alpha = 0.5)
ax.fill_between(x, 2, y1, color=colors[10], alpha = 0.5)
ax.fill_between(x, 1, 0, color=colors[4], alpha = 0.5)

ax.text(
    0.30, 
    0.90, 
    r"Localized", 
    transform=ax.transAxes,
    fontsize=16,
    color = 'black',
    horizontalalignment="center",
    verticalalignment="center",
)

ax.text(
    0.70, 
    0.65, 
    r"Non-Ergodic-Extended", 
    transform=ax.transAxes,
    fontsize=16,
    color = 'black',
    horizontalalignment="center",
    verticalalignment="center",
)

ax.text(
    0.30, 
    0.30, 
    r"Ergodic", 
    transform=ax.transAxes,
    fontsize=16,
    color = 'black',
    horizontalalignment="center",
    verticalalignment="center",
)

ax.minorticks_on()
ax.tick_params(axis='both', which='major', labelsize=12)

ax.set_xlabel(r"$\mu$", fontsize=15)
ax.set_ylabel(r"$\gamma$", fontsize=15)

ax.set_ylim(0,2)
ax.set_xlim(1,2)
ax.set_xticks(np.arange(1,2.1,0.5))
ax.set_yticks(np.arange(0,2.1,0.5))
ax.set_xticklabels(np.arange(1,2.1,0.5))
ax.set_yticklabels(np.arange(0,2.1,0.5))


plt.savefig("phase_diagram.jpg",bbox_inches='tight', pad_inches=0.1)
plt.show()    
```
